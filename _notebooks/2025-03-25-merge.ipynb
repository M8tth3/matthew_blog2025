## Popcorn Hack


```java
public class MergeSortedArrays {

    public static int[] mergeArrays(int[] array1, int[] array2) {
        int n1 = array1.length;
        int n2 = array2.length;
        int[] mergedArray = new int[n1 + n2];

        int i = 0, j = 0, k = 0;

        // Merge the arrays
        while (i < n1 && j < n2) {
            if (array1[i] <= array2[j]) {
                mergedArray[k] = array1[i];
                i++;
            } else {
                mergedArray[k] = array2[j];
                j++;
            }
            k++;
        }

        // Copy remaining elements of array1, if any
        while (i < n1) {
            mergedArray[k] = array1[i];
            i++;
            k++;
        }

        // Copy remaining elements of array2, if any
        while (j < n2) {
            mergedArray[k] = array2[j];
            j++;
            k++;
        }

        return mergedArray;
    }

    public static void main() {
        int[] array1 = {1, 3, 5, 7};
        int[] array2 = {2, 4, 6, 8};

        int[] mergedArray = mergeArrays(array1, array2);

        System.out.println("Merged Array:");
        for (int num : mergedArray) {
            System.out.print(num + " ");
        }
    }
}
MergeSortedArrays.main();
```

### Time Complexity of Merge Sort

- **Best, Worst, and Average Case:** O(n log n)

### Comparison with Bubble Sort and Quicksort

- **Bubble Sort:** 
    - **Best Case:** O(n)
    - **Worst and Average Case:** O(n^2)
    - Generally not preferred for large datasets.

- **Quicksort:**
    - **Best and Average Case:** O(n log n)
    - **Worst Case:** O(n^2)
    - Faster than merge sort but can degrade to O(n^2).

### When to Prefer Merge Sort

- Stability is required.
- Data is too large to fit into memory.

### Divide and Conquer Algorithm

Merge sort:
1. **Divides** the array.
2. **Conquers** each half by sorting.
3. **Combines** the sorted halves.

### Stability of Merge Sort

Merge sort is stable, preserving the relative order of equal elements.


```java
public class MergeSort {

    public static void mergeSort(int[] array, int left, int right) {
        if (left < right) {
            int middle = (left + right) / 2;

            mergeSort(array, left, middle);
            mergeSort(array, middle + 1, right);

            merge(array, left, middle, right);
        }
    }

    private static void merge(int[] array, int left, int middle, int right) {
        int n1 = middle - left + 1;
        int n2 = right - middle;

        int[] leftArray = new int[n1];
        int[] rightArray = new int[n2];

        System.arraycopy(array, left, leftArray, 0, n1);
        System.arraycopy(array, middle + 1, rightArray, 0, n2);

        int i = 0, j = 0;
        int k = left;

        while (i < n1 && j < n2) {
            if (leftArray[i] <= rightArray[j]) {
                array[k] = leftArray[i];
                i++;
            } else {
                array[k] = rightArray[j];
                j++;
            }
            k++;
        }

        while (i < n1) {
            array[k] = leftArray[i];
            i++;
            k++;
        }

        while (j < n2) {
            array[k] = rightArray[j];
            j++;
            k++;
        }
    }

    public static int mergeSortWithComparisons(int[] array, int left, int right) {
        int comparisons = 0;
        if (left < right) {
            int middle = (left + right) / 2;

            comparisons += mergeSortWithComparisons(array, left, middle);
            comparisons += mergeSortWithComparisons(array, middle + 1, right);

            comparisons += mergeWithComparisons(array, left, middle, right);
        }
        return comparisons;
    }

    private static int mergeWithComparisons(int[] array, int left, int middle, int right) {
        int comparisons = 0;
        int n1 = middle - left + 1;
        int n2 = right - middle;

        int[] leftArray = new int[n1];
        int[] rightArray = new int[n2];

        System.arraycopy(array, left, leftArray, 0, n1);
        System.arraycopy(array, middle + 1, rightArray, 0, n2);

        int i = 0, j = 0;
        int k = left;

        while (i < n1 && j < n2) {
            comparisons++;
            if (leftArray[i] <= rightArray[j]) {
                array[k] = leftArray[i];
                i++;
            } else {
                array[k] = rightArray[j];
                j++;
            }
            k++;
        }

        while (i < n1) {
            array[k] = leftArray[i];
            i++;
            k++;
        }

        while (j < n2) {
            array[k] = rightArray[j];
            j++;
            k++;
        }

        return comparisons;
    }

    public static void main() {
        int[] array = {12, 11, 13, 5, 6, 7};
        System.out.println("Given Array");
        for (int i : array) {
            System.out.print(i + " ");
        }
        System.out.println();

        mergeSort(array, 0, array.length - 1);

        System.out.println("\nSorted array");
        for (int i : array) {
            System.out.print(i + " ");
        }
        System.out.println();

        int[] arrayWithComparisons = {12, 11, 13, 5, 6, 7};
        int comparisons = mergeSortWithComparisons(arrayWithComparisons, 0, arrayWithComparisons.length - 1);

        System.out.println("\nSorted array with comparisons");
        for (int i : arrayWithComparisons) {
            System.out.print(i + " ");
        }
        System.out.println("\nNumber of comparisons: " + comparisons);
    }
}
MergeSort.main();
```
